# Notebook for creating quality metric figure (Figure 3)

```python
# Configutation part #

# Output of Pycasso analysis.ipynb
scad_file = './lr/final_output_scad/TableS19.csv'
# Output of Pycasso analysis.ipynb
mcp_file = './lr/final_output_mcp/TableS20.csv'
# Output of Pycasso analysis.ipynb
l1_file = './lr/final_output_l1/TableS21.csv'
# Output of ABESS analysis.ipynb
abess_file = './abess/final_output_abess/TableS17.csv'
# Output of HHS analysis.ipynb
hhs_file = './hhs/final_output_hhs/TableS18.csv'
# Output of Elastic net analysis.ipynb
elastic_net_file = './elastic_net/final_output_elastic_net/TableS22.csv'
# Output of Elastic net analysis.ipynb
who_file = './who_pred/TableS23.csv'

######################
```

```python
import numpy as np
import pandas as pd

from matplotlib import pyplot as plt
```

```python
drugs = ['Rifampicin',
        'Isoniazid',
        'Pyrazinamide',
        'Ethambutol',
        'Streptomycin',
        'Kanamycin',
        'Amikacin',
        'Capreomycin',
        'Ofloxacin',
        'Moxifloxacin',
        'Ciprofloxacin',
        'Ethionamide',
        'Prothionamide']
```

```python
abess_result = pd.read_csv(abess_file, sep='\t', index_col=0, header=0)
hhs_result = pd.read_csv(hhs_file, sep='\t', index_col=0, header=0)
scad_result = pd.read_csv(scad_file, sep='\t', index_col=0, header=0)
mcp_result = pd.read_csv(mcp_file, sep='\t', index_col=0, header=0)
l1_result = pd.read_csv(l1_file, sep='\t', index_col=0, header=0)
elastic_result = pd.read_csv(elastic_net_file, sep='\t', index_col=0, header=0)
who_result = pd.read_csv(who_file, sep='\t', index_col=0, header=0)
```

```python
result = pd.DataFrame(index=drugs)
result['scad'] = scad_result['auc']
result['mcp'] = mcp_result['auc']
result['l1'] = l1_result['auc']
result['elastic'] = elastic_result['auc']
result['abess'] = abess_result['mean_auc']
result['hhs'] = hhs_result['auc']
result['WHO'] = who_result['auc']
```

```python
result
```

```python
fig = result.plot(kind='bar', figsize=(15,8)).get_figure()
plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
fig.savefig('Fig3.png')
```
